```python
import requests as rs
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
house_name =[]
city_name =[]
price =[]
area_sq_ft =[]
status = []
for i in range(1, 244):
    url="https://www.makaan.com/hyderabad-residential-property/rent-property-in-hyderabad-city?budget=5000,&page=244&_=1699870743812"
    header_requests={"Accept-Encoding":"gzip, deflate, br","User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"}
    response = rs.get(url, headers=header_requests)
    html_code = response.text
    soup = BeautifulSoup(html_code)
    titles = soup.find_all('div', attrs={'class' : 'title-line'})
    for title in titles:
        house_name.append(title.text)
    titles = soup.find_all('div', attrs={'class' : 'locWrap'})
    for title in titles:
        city_name.append(title.text)
    titles = soup.find_all('div', attrs={'data-type' :"price-link"})
    for title in titles:
        price.append(title.text)     
    titles = soup.find_all('td', attrs={'class' :"size"})
    for title in titles:
        area_sq_ft.append(title.text)    
    titles = soup.find_all('td', attrs={'class' :"val"})
    for title in titles:
        status.append(title.text)
```

```python
house_name =[]
city_name =[]
price =[]
area_sq_ft =[]
status = []
page = 244
for i in range(1,page):
    url="https://www.makaan.com/hyderabad-residential-property/rent-property-in-hyderabad-city?budget=5000,&page=244&_=1699870743812".format(i)
    header_requests={"Accept-Encoding":"gzip, deflate, br","User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"}
    response = rs.get(url, headers=header_requests)
    html_code = response.text
    soup = BeautifulSoup(html_code)
    container = soup.find_all('div', attrs={'class' :"cardLayout clearfix"})
    for containers in container:
            house = containers.find('div', attrs={'class' : "title-line"})
            if house is None:
                house_name.append(np.NAN)
            else:
                house_name.append(house.text)
            city = containers.find('div', attrs={'class' : "locWrap"})
            if city is None:
                city_name.append(np.NAN)
            else:
                city_name.append(city.text)
            Cost = containers.find('div', attrs={'date-type' : "price-link"})
            if Cost is None:
                price.append(np.NAN)
            else:
                price.append(Cost.text)
            area = containers.find('td', attrs={'class' : "size"})
            if area is None:
                area_sq_ft.append(np.NAN)
            else:
                area_sq_ft.append(area.text)
            house = containers.find('td', attrs={'class' : "val"})
            if house is None:
                status.append(np.NAN)
            else:
                status.append(house.text)
```

```python
df=pd.DataFrame({"house_name":house_name,
                 "city_name":city_name,
                "price": price,
                "area_sq_ft":area_sq_ft,
                "status":status})
```
