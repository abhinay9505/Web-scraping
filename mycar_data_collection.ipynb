```python
import requests as rs
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re


# In[2]:


car_name =[]
price =[]
ower = []
hybrid_petrol =[]
km =[]
date = []
eml_month =[]
ref =[]

page = 121
for i in range(1,page+1):
    #url
    url = f"https://www.mycar.mu/car/buy?page={i}"
    #requests
    respone = rs.get(url)
    html = respone.text
    #typecasting
    soup = BeautifulSoup(html)
    container = soup.find_all("div",attrs={"class":"col-sm-6 col-md-4 d-flex"})
    for containers in container:
        car_names = containers.find("span",attrs={"itemprop":"name"})
        if car_names is None:
            car_name.append(np.NAN)
        else:
            car_name.append(car_names.text)
        prices = containers.find("span",attrs={"class":"price"})
        if prices is None:
            price.append(np.NAN)
        else:
            price.append(prices.text)
        owers = containers.find("span",attrs={"itemprop":"itemCondition"})
        if owers is None:
            ower.append(np.NAN)
        else:
            ower.append(owers.text)
        petrol = containers.find("span",attrs={"itemprop":"fuelType"})
        if petrol is None:
            hybrid_petrol.append(np.NAN)
        else:
            hybrid_petrol.append(petrol.text)
        kms = containers.find("span",attrs={"itemprop":"value"})
        if kms is None:
            km.append(np.nan)
        else:
            km.append(kms.text)
        eml = containers.find("span",attrs={"class":"price-per-month"})
        if eml is None:
            eml_month.append(np.NAN)
        else:
            eml_month.append(eml.text)
        refe = containers.find("div",attrs={"class":"title-wrap ref position-relative"})
        if refe is None:
            ref.append(np.nan)
        else:
            ref.append(refe.text)
    for i in soup.find_all("div",class_="col-sm-6 col-md-4 d-flex"):
        regex= re.findall("[A-Z][a-z]{2}\s?\d{4}",i.text.strip())
        if regex:
            date.append(regex[0])
        else:
            date.append(np.nan)
```

```python
df =pd.DataFrame({"car_name":car_name,
                 "price":price,"ower":ower,"hybrid_petrol":hybrid_petrol,"km":km
,"eml_month":eml_month,
"ref":ref,"date":date })
```
